# A Single Neuron:

Example #1:

```python
#The input here is a list of 3 features (1 sample).
inputs = [1, 2, 3]

#Each connection between neurons has a weight associated with it, which is a trainable factor of how much of this input to use
weights = [0.2, 0.8, -0.5]

#The purpose of the bias is to offset the output positively or negatively
bias = 2

# the hypotheses H(x) = W*X + bias
outputs = inputs[0]*weights[0] + inputs[1]*weights[1] + inputs[2]*weights[2] + bias

print(outputs)
```

```python
#Solve the problem using Numpy
# firstly we will import the needed library.
import numpy as np

inputs = [1, 2, 3]

weights = [0.2, 0.8, -0.5]

bias = 2

# insted of this: outputs = inputs[0]*weights[0] + inputs[1]*weights[1] + inputs[2]*weights[2] + bias, we use dot function,
# to multiply the vectors:
outputs = np.dot(inputs, np.array(weights).T) + bias

print(outputs)
```

# A Layer of Neurons

Example #3:

here there are 4 inputs in the input layer. and 3 neurons in the output layer.

```python
inputs = [ 1, 2, 3, 2.5 ]

# each neuron has a list of weights:
weights1 = [ 0.2, 0.8, - 0.5, 1 ]
weights2 = [ 0.5, - 0.91, 0.26, - 0.5 ]
weights3 = [ - 0.26, - 0.27, 0.17, 0.87 ]

# each neuron has a bias:
bias1 = 2
bias2 = 3
bias3 = 0.5
```

```python
# here the output are 3 neurons, so it will be a list of three values.
outputs = [
                # Neuron 1:
                inputs[0] * weights1[0] + inputs[1] * weights1[1] + inputs[2] * weights1[2] + inputs[3] * weights1[3] + bias1,
    
                # Neuron 2:
                inputs[0] * weights2[0] + inputs[1] * weights2[1] + inputs[2] * weights2[2] + inputs[3] * weights2[3] + bias2,
    
                # Neuron 3:
                inputs[0] * weights3[0] + inputs[1] * weights3[1] + inputs[2] * weights3[2] + inputs[3] * weights3[3] + bias3
           ]
```

```python
outputs
```

```python
#Solve the problem using Numpy
inputs = [ 1, 2, 3, 2.5 ]

''' insted of these three lists:
weights1 = [ 0.2, 0.8, - 0.5, 1 ]
weights2 = [ 0.5, - 0.91, 0.26, - 0.5 ]
weights3 = [ - 0.26, - 0.27, 0.17, 0.87 ]

we will put them together in one array:
'''
weights = [
    [ 0.2 , 0.8 , - 0.5 , 1 ],
    [ 0.5 , - 0.91 , 0.26 , - 0.5 ],
    [ - 0.26 , - 0.27 , 0.17 , 0.87 ]
          ]

# and list for the biases values:
biases = [ 2.0 , 3.0 , 0.5 ]
# here the output are 3 neurons, so it will be a list of three values.
layer_outputs = np.dot(inputs, np.array(weights).T) + biases

print(layer_outputs)
```

# A Batch of Data

Often, neural networks expect to take in many samples at a time for two reasons. One reason
is that its faster to train in batches in parallel processing, 
and the other reason is that batches help with generalization during training.

If you fit (perform a step of a training process) on one
sample at a time, youre highly likely to keep fitting to that individual sample, rather than
slowly producing general tweaks to weights and biases that fit the entire dataset. Fitting or
training in batches gives you a higher chance of making more meaningful changes to weights
and biases.

An example of a batch of data could look like:

For example, every sample is a describing for a house.

So:
X1: the area of this house.
X2: number of rooms.
..
.. And so on.


```python
import numpy as np

# Batch of 3 examples:
inputs = [
    [ 1.0 , 2.0 , 3.0 , 2.5 ],
    [ 2.0 , 5.0 , - 1.0 , 2.0 ],
    [ - 1.5 , 2.7 , 3.3 , - 0.8 ]
         ]
weights = [
    [ 0.2 , 0.8 , - 0.5 , 1.0 ],
    [ 0.5 , - 0.91 , 0.26 , - 0.5 ],
    [ - 0.26 , - 0.27 , 0.17 , 0.87 ]
          ]
biases = [ 2.0 , 3.0 , 0.5 ]

outputs = np.dot(inputs, np.array(weights).T) + biases
print(outputs)
```

As you can see, our neural network takes in a group of samples (inputs) and outputs a group of
predictions.

# Why we use W.T (Transported of weights array)?

we need to perform the dot
product of each input and each weight set in all of their combinations. The dot product takes the
row from the first array and the column from the second one, but currently the data in both arrays
are row-aligned. Transposing the second array shapes the data to be column-aligned. The matrix
product of inputs and transposed weights will result in a matrix containing all atomic dot products
that we need to calculate. The resulting matrix consists of outputs of all neurons after operations
performed on each input sample:

```python
def relu(x):
    """
    ReLU (Rectified Linear Unit) activation function
    
    Args:
        x: Input array (numpy array or scalar)
    
    Returns:
        Element-wise ReLU activation: max(0, x)
    """
    return np.maximum(0, x)
```

```python

def sigmoid(x):
    """
    Sigmoid activation function
    
    Args:
        x: Input array (numpy array or scalar)
    
    Returns:
        Element-wise sigmoid activation: 1 / (1 + exp(-x))
    """
    return 1 / (1 + np.exp(-x)) 
```

```python
test_inputs = np.array([-2, -1, 0, 1, 2])
print("Input:", test_inputs)
print("ReLU:", relu(test_inputs))
print("Sigmoid:", sigmoid(test_inputs))
```
