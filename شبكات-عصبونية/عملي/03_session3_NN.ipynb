#  Architecture of the Neural Network

### Our neural network will have the following structure:

* Input Layer: 1 layer with 30 nodes (representing the 5x6 grid).
* Hidden Layer: 1 layer with 5 nodes.
* Output Layer: 1 layer with 3 nodes (representing the letters A, B, and C).



```python
# activation function
def sigmoid(x):
    #Write your code here
	return 1 / (1 + np.exp(-x))
```

```python
# initializing the weights randomly
w1 = np.random.randn(5, 30)#Write your code here
w2 = np.random.randn(3, 5)#Write your code here
```

```python
# Creating the Feed forward neural network
def f_forward(x, w1, w2):
	# hidden
	o1 = np.dot(x, w1.T) 
	z1 =  sigmoid(o1)
	o2 = np.dot(z1, w2.T) 
	z2 = sigmoid(o2)
	return z2
```

```python
res = f_forward(x, w1, w2)
res
```
